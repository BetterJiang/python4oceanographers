I just had my [import anti-gravity feeling](http://xkcd.com/353/) feeling
when I found out about [folium.](https://folium.readthedocs.org/en/latest/)

I have no knowledge of JavaScript nor [leaflet](http://leafletjs.com/), yet
with just a few lines of python code I could generate a plot with all the HF
radar station positions on an interactive map.

To start you have to download the stations positions
[kml file](http://hfrnet.ucsd.edu/rtv/ss.php) and extract the lon, lat and name
data of each station.

```python
from fastkml.kml import KML

def read_kml(fname='ss.kml'):
    kml = KML()
    kml.from_string(open(fname).read())
    points = dict()
    for feature in kml.features():
        for placemark in feature.features():
            if placemark.styleUrl.startswith('#hf'):
                points.update({placemark.name:
                            (placemark.geometry.y, placemark.geometry.x, )})
    return points
```

Then create a dictionary with data information.

```python
fname = './data/ss.kml'
locations = read_kml(fname)
```

Now just choose a starting point and a zoom level for your map, and loop
through pass the dictionary.

```python
import folium

folium.initialize_notebook()

from IPython.display import IFrame


def inline_map(m):
    """Takes a folium instance or a html path and load into an iframe."""
    if isinstance(m, folium.Map):
        m._build_map()
        srcdoc = m.HTML.replace('"', '&quot;')
        embed = HTML('<iframe srcdoc="{srcdoc}" '
                     'style="width: 100%; height: 500px; '
                     'border: none"></iframe>'.format(srcdoc=srcdoc))
    elif isinstance(m, (str, unicode)):
        embed = IFrame(m, width=750, height=500)
    return embed


radars = folium.Map(location=[40, -122], zoom_start=5,
                    tiles='OpenStreetMap', width=500, height=500)

for name, location in locations.items():
    radars.simple_marker(location=location, popup=name)

radars.create_map('map.html')
```

```python
inline_map(radars)
```

That's it, now you have an interative map!  (Try clicking at the stations.)
