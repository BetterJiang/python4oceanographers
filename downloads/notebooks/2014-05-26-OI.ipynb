```python
frac = 0.05  # Fraction of data to retain for mapping.
e = 0.75  # Random error to add to obs (in obs units, here deg C).
dx = dy = 1/3
lons = np.arange(-78, -61+dx, dx)
lats = np.arange(31, 42+dy, dy)
xg, yg = np.meshgrid(lons, lats)  # Regular grid.
```

```python
merc = loadmat('./data/mercator_temperature.mat', squeeze_me=True)
```

```python
x, y = merc['x'], merc['y']
X, Y = np.meshgrid(x, y)

level = 10  # 10 meters SST.
data = ma.masked_invalid(merc['temp'][level, ...])

#ax
#depth
#lat
#corners
#lon
#f
#temp
```

```python
def make_map(llcrnrlon=-78, urcrnrlon=-61,
             llcrnrlat=31, urcrnrlat=42,
             projection='merc', resolution='i'):
    m = Basemap(llcrnrlon=llcrnrlon, urcrnrlon=urcrnrlon,
                llcrnrlat=llcrnrlat, urcrnrlat=urcrnrlat,
                projection=projection, resolution=resolution,
                lat_ts=(llcrnrlat+urcrnrlat)/2)

    fig, ax = plt.subplots(figsize=(6, 4), facecolor='w')
    m.ax = ax
    m.fillcontinents()
    dx = 3
    dy = 2
    meridians = np.arange(np.fix(llcrnrlon),
                          np.fix(urcrnrlon) + dx, dx)
    parallels = np.arange(np.fix(llcrnrlat),
                          np.fix(urcrnrlat) + dy, dy)
    kw = dict(linewidth=0)
    m.drawparallels(parallels, labels=[1, 0, 0, 0], **kw)
    m.drawmeridians(meridians, labels=[0, 0, 0, 1], **kw)
    return fig, m
```

```python
llcrnrlon, urcrnrlon = -78, -61
llcrnrlat, urcrnrlat = 31, 42
    
fig, m = make_map()
m.pcolormesh(X, Y, data, latlon=True)
_ = m.ax.set_title('Temperature at %s m'
                   'from Mercator analysis 2004-Mar-01' % level)
```

Get distances separating every pair of stations using approximate
spherical calculation

```python
rearth = 6370800  # Meters.
```

xkm, ykm will be the coordinates of the data (converted from lon/lat)

```python
ykm = rearth * (Y - llcrnrlat) * np.pi / 180 / 1000
xkm = (rearth * (np.pi/180 * (X - llcrnrlon)) *
       np.cos(np.pi / 180 * 0.5 * (Y + llcrnrlat)) /1000)
```

Build matrices that repeat columns of x and rows of y. Then X-X' and Y-Y' will be all possible distance combinations between pairs of points

```python
X = np.tile(xkm[:, None], [1, ykm.size])
Y = np.tile(ykm, (xkm.size, 1))
Rdd = np.sqrt((X-X.T)**2 + (Y-Y.T)**2)
```

triu sets the lower triangle to zeros.  We don't want those points because they are duplicates of the upper triangle (xi-xj) = -(xj-xi)

```python
Rddu = np.triu(Rdd)
```

Add some random noise to data

```python
print('Adding random noise of +/- %s ℃' % e)
print('or variance %s ℃ squared' % e**2)
d += e * np.random.randn(d.size)
```

Compute the binnned lagged covariance function.
Remove mean before calculating covariance.

```python
d -= d.mean()  # Remove the mean.
C = np.dot(d[:, None], d[:, None].T)  # Covariance.

dr = 30 
r = np.r_[0, np.arange(0.5*dr, 1000, dr)]
cf = np.zeros_like(r)
```

```python
cf[0] = np.diag(C).mean()
for k in range(1, len(r)):
    mask = Rddu(:) > r(i)-0.5*dr & Rddu(:)<=r(i)+0.5*dr
    cf[k] = C[mask].mean()
```

```python
cf[0] = mean(diag(C));

for i=2:length(r)
  tmp = find(Rddu(:)>r(i)-0.5*dr & Rddu(:)<=r(i)+0.5*dr);
  cf(i) = mean(C(tmp));
end
```
