```python
import numpy as np


def fake_tide(t, M2amp, M2phase):
    """
    Generate a minimally realistic-looking fake semidiurnal tide.
    
    t is time in hours
    phases are in radians
    """
    out = M2amp * np.sin(2 * np.pi * t / 12.42 - M2phase)
    return out
```

```python
N = 500
res = 5
noise = np.random.randn(N)

t = np.arange(N)

amp = 2
phase = 0

u = fake_tide(t, amp, phase) + res + noise
```

```python
%matplotlib inline
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(11, 2.75))
ax.plot(t, u, 'darkgreen', label='sinal')
leg = ax.legend()
```

```python
mean = np.mean(u)

std = amp * np.std(u) / (amp**0.5)

phase = 0

# First guess.
first_guess = fake_tide(t, std, 0.25) + mean  # Fase está um pouco fora de propósito!

fig, ax = plt.subplots(figsize=(11, 2.75))
ax.plot(t, first_guess, 'darkorange', label='chute inicial')
ax.plot(t, u, 'darkgreen', label='sinal')
leg = ax.legend()
```

```python
from scipy.optimize import leastsq  # Least-squares.

# x[0] -> chute amp, x[1] -> chute phase, x[2] -> média
# Otimizando contra o sinal "u".
optimize_func = lambda x: x[0] * np.sin(2 * np.pi * t / 12.42 + x[1]) + x[2] - u

est_std, est_phase, est_mean = leastsq(optimize_func, [std, phase, mean])[0]

# Recriar o sinal com os valores estimados.
fit = fake_tide(t, est_std, est_phase) + est_mean
```

```python
fig, ax = plt.subplots(figsize=(11, 2.75))
ax.plot(t, fit, 'darkorange', label='estimated')
ax.plot(t, first_guess, 'crimson', label='guess')
ax.plot(t, u, 'darkgreen', label='sinal')
leg = ax.legend()
ax.set_xlim(100, 150)
```
