```python
import iris

url = 'http://data.nodc.noaa.gov/thredds/dodsC/woa/WOA09/NetCDFdata/silicate_annual_5deg.nc'

cubes = iris.load_raw(url)
print(cubes)
```

```python
%matplotlib inline

import numpy as np
import numpy.ma as ma
import iris.plot as iplt
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

cube = cubes[0]
surface = cube[0, 0, ...]

surface.data = ma.log10(ma.masked_equal(surface.data, 0))

fig, ax = plt.subplots(figsize=(15, 9),
                       subplot_kw=dict(projection=ccrs.PlateCarree()))
ax.coastlines()
iplt.outline(surface)
cs = iplt.pcolormesh(surface)
pt = iplt.points(surface, alpha=0.5)
gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER

cbar = fig.colorbar(cs, orientation='vertical', extend='both',
                    shrink=0.7)
tick_levels = [0, 1, 2, 3, 4]
cbar.set_ticks(tick_levels)
cbar.set_label('Number of obs [log10]')
```

```python
from iris.experimental.raster import export_geotiff


export_geotiff(surface, 'surface.tiff')
```
