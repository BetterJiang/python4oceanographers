```python
import pyepsg

from osgeo import osr
from osgeo import gdal

import cartopy.crs as ccrs

import matplotlib.pyplot as plt

gdal.UseExceptions()


def geotiff2cartopy(fname):
    ds = gdal.Open(fname)  # Read with GDAL.
    data = ds.ReadAsArray()  # Image data array.
    gt = ds.GetGeoTransform()
    # Get projection.
    proj = ds.GetProjection()
    inproj = osr.SpatialReference()
    inproj.ImportFromWkt(proj)
    if False:  # If using basemap (or https://github.com/SciTools/cartopy/issues/153).
        proj = inproj.ExportToProj4()
    # Convert to cartopy projection.
    projcs = inproj.GetAuthorityCode('PROJCS')
    geogcs = inproj.GetAuthorityCode('GEOGCS')
    if projcs:
        projection = ccrs.epsg(projcs)
    else:
        projection = ccrs.PlateCarree()
    # Cartopy figure.
    subplot_kw = dict(projection=projection)
    fig, ax = plt.subplots(subplot_kw=subplot_kw)
    extent = (gt[0], gt[0] + ds.RasterXSize * gt[1],
              gt[3] + ds.RasterYSize * gt[5], gt[3])
    img = ax.imshow(data[:3, :, :].transpose((1, 2, 0)), extent=extent, origin='upper')
    return fig, ax
```

```python
fig, ax = geotiff2cartopy('./data/manhattan.tif')
```

```python
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER


fig, ax = geotiff2cartopy('data/BlueMarbleNG-TB_2004-12-01_rgb_720x360.tiff')

gl = ax.gridlines(draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
```
